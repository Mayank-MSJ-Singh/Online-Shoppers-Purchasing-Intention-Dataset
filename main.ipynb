```python
print(pd.unique(df["VisitorType"]))
print(pd.unique(df["Month"]))
```

```python
Returning_Visitor = 0
New_Visitor = 0
Other = 0

for i in range(0, len(df)):
    if df["VisitorType"][i] == "Returning_Visitor":
        Returning_Visitor += 1
    elif df["VisitorType"][i] == "New_Visitor":
        New_Visitor += 1
    else:
        Other += 1
```

```python
percent = lambda x: ((x/len(df))*100)
```

```python
Returning_Visitor_Percent = round(percent(Returning_Visitor), 2)
New_Visitor_Percent = round(percent(New_Visitor), 2)
Other_Percent = round(percent(Other), 2)
```

```python
VT_pie = [Returning_Visitor, New_Visitor, Other]
VT_pie_labels = ["Returning_Visitor  "+ str(Returning_Visitor_Percent), "New_Visitor  "+str(New_Visitor_Percent), "Other  "+str(Other_Percent)]
print(plt.pie(VT_pie, labels=VT_pie_labels))
```
